```python
import pickle
import pandas as pd
import numpy as np
```

```python
all_reviews = []
```

## Loading PeoplePerHour files


```python
for i in range(1,7):
    file_name = "data_oct4_"+str(i)+".pkl"
    pickle_in = open(file_name,"rb")
    reviews = pickle.load(pickle_in)
    all_reviews += reviews
#     concat_df = new_df.append(df, ignore_index=True)
```

```python
len(all_reviews)
```

```python
all_reviews[0]
```

```python
new_review_dict = {"user_no":[],"user_url":[],"reviews":[]}
```

```python
for i in range(len(all_reviews)):
    row = all_reviews[i]
    for review in row["reviews"]:
        new_review_dict["user_no"].append(row["user_no"])
        new_review_dict["user_url"].append(row["user_url"])
        new_review_dict["reviews"].append(review)
```

```python
PeoplePerHour_df = pd.DataFrame.from_dict(new_review_dict)
```

## Loading 99Design files

```python
all_reviews = []

for i in range(1,3):
    file_name = "data_oct4_Art_Illustration_"+str(i)+".pkl"
    pickle_in = open(file_name,"rb")
    reviews = pickle.load(pickle_in)
    all_reviews += reviews
```

```python
all_reviews[0]
```

```python
new_review_dict = {"user_no":[],"user_url":[],"reviews":[]}
```

```python
for i in range(len(all_reviews)):
    row = all_reviews[i]
    for review in row["reviews"]:
        new_review_dict["user_no"].append(row["user_no"])
        new_review_dict["user_url"].append(row["user_url"])
        new_review_dict["reviews"].append(review[0][1:-1])
```

```python
design_99_df = pd.DataFrame.from_dict(new_review_dict)
```
